# GENRES

```python
genres_df = dict(movies['Genres'].value_counts())
genres = set()
for g,count in genres_df.items():
    x_g = g.split(",")
    for xg in x_g:
        if xg in genres:
            continue
        else:
            genres.add(xg)
```

```python
genres_count = dict()
for g,count in genres_df.items():
    x_g = g.split(",")
    for xg in x_g:
        if xg in genres_count.keys():
            genres_count[xg] = genres_count.get(xg)+1
        else:
            genres_count[xg] = 1
```

```python
genres_count_df = pd.DataFrame(genres_count.items(), columns=['Genre', 'Count'])
```

```python
genres_count_df.sort_values(ascending=False, by='Count').iplot(kind="bar", x='Genre', xTitle='Genre', yTitle='Count', title='Count of Genres', color='blue')
```

```python
plt.subplots(figsize = (10,10))

wordcloud = WordCloud (background_color = 'white', width = 720, height = 720).generate(' '.join(genres_count_df['Genre']))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
