```python
import pandas as pd
import plotly.express as px
```

```python
data_df = pd.read_csv("https://nw-bc-project3.s3.us-east-2.amazonaws.com/cleaned_data.csv")
```

```python
data_df
```

```python
data_df = data_df.rename(columns={"Disney+": "Disney"})
```

```python
data_df = px.data.gapminder().query("Disney == 1").query("Top_Genres")
fig = px.pie(data_df, names='Top_Genres', title='Movie Genres on Disney+')
fig.show()
```
