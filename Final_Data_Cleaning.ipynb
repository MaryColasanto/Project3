```python
import pandas as pd 
import numpy as np
```

```python
# Importing Data

netflix_df = pd.read_csv("Resources/netflix.csv")
imdb_df = pd.read_csv("Resources/imdb.csv")
tom_df = pd.read_csv("Resources/rottentomatoes.csv")
```

# Cleaning up Netflix Data

```python
netflix_df.head()
```

```python
netflix_df = netflix_df[["Title", "Year", "Age", "IMDb", "Rotten Tomatoes", 
                             "Netflix", "Hulu", "Prime Video", "Disney+", "Directors", "Genres", "Country", "Runtime"]]
netflix_df.head()
```

```python
netflix_df.columns
```

# Cleaning up IMDB Data

```python
imdb_df.head()
```

```python
imdb_df.columns
```

```python
imdb_df = imdb_df[["title", "year", "production_company"]]
```

```python
imdb_df = imdb_df.rename(columns={"title": "Title", "year": "Year"})
imdb_df.head()
```

Cleaning Rotten Tomatoes

```python
tom_df.head()
```

```python
tom_df = tom_df[["movie_title", "original_release_date", "production_company"]]
tom_df
```

```python
# tom_df = tom_df.rename(columns={"movie_title": "Title"})
# tom_df = tom_df.rename(columns={"original_release_date": "Year"})
tom_df = tom_df.rename(columns={"movie_title": "Title", "original_release_date": "Year"})
```

```python
tom_df.head()
```

```python
tom_df['Year'] = pd.DatetimeIndex(tom_df['Year']).year
tom_df["Year"] = tom_df["Year"].astype("object")
tom_df.head()
```

# Merging Datasets

```python
imdb_tom_data = pd.merge(imdb_df, tom_df, on=['Title', 'Year'], how="left")
imdb_tom_data.head()
```

```python
imdb_tom_data
```

```python
combined_data = pd.merge(netflix_df, imdb_df, on=['Title', 'Year'], how="left")
combined_data.head()
```

```python
combined_data["production_company"].isna().value_counts()
```

```python
combined_tom_data = pd.merge(combined_data, tom_df, on=['Title', 'Year'], how="left")
combined_tom_data.head()
```

```python
combined_tom_data["production_company_x"].isna().value_counts()
```

```python
combined_tom_data["production_company_y"].isna().value_counts()
```

```python
countries = combined_data['Country'].str.split(',', expand=True)
countries
```

```python
countries[0]
```

```python
combined_data["Country"] = countries[0]
combined_data.head()
```

```python
combined_data.dtypes
```

```python
no_percent = combined_data["Rotten Tomatoes"].str.replace('%', '')
```

```python
no_percent = no_percent.dropna()
```

```python
no_percent = no_percent.astype(int)
```

```python
no_percent = no_percent / 10
```

```python
no_percent
```

```python
combined_data["Rotten Tomatoes"] = no_percent
combined_data.head()
```

```python
genres = combined_data['Genres'].str.split(',', expand=True)
genres
```

```python
combined_data["Top_Genres"] = genres[0]
combined_data.head()
```

```python
combined_data["Age"].unique()
```

```python
directors = combined_data['Directors'].str.split(',', expand=True)
directors
```

```python
combined_data["Top_Director"] = directors[0]
combined_data.head()
```

```python
combined_data["Top_Director"].isna().value_counts()
```

```python
combined_data.dtypes
```

```python
combined_data.to_csv("Resources/cleaned_data.csv", encoding="utf-8", index=False)
```
